## Libraries

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import pearsonr
```

## Data

```python
raw_df = pd.read_csv('stocks.csv')
```

```python
raw_df.info()
```

```python
df = raw_df.copy()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Open price for each ticker

```python
plt.title('Open price trends for 4 companies', pad=15)
for tick in df['Ticker'].unique():
    tick_df = df[df['Ticker']==tick]
    plt.plot(tick_df['Date'], tick_df['Open'], label=tick)
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

*Apple and Google show lower daily fluctuations and overall too. Whereas, Netflix and Microsoft show higer fluctuations.*

## Hypothesis: Apple, Microsoft and Google show similar open price movements

```python
### correlation and p-value check on open price
corr_dict = dict()
tickers = df['Ticker'].unique()
for tick in tickers:
    corr_dict[tick] = df[df['Ticker']==tick]['Open'].tolist()
corr_df = pd.DataFrame(corr_dict)

print('Ticker1   Ticker2   Correlation   p-value   statistically significant:')
print()
for i in range(len(tickers)-1):
    tick1 = tickers[i]
    for j in range(i+1, len(tickers)):
        tick2 = tickers[j]
        corr, p_v = pearsonr(corr_df[tick1], corr_df[tick2])
        print('{0}   {1}   {2}   {3}   {4}'.format(tick1, tick2, corr, p_v, p_v<0.05))
```

*The results are statistically significant.*
<br>**Therefore, Google, Microsoft and Apple show similar open price movements in market. Apple and Microsoft show significant correlation.**

## Volume

```python
for tick in df['Ticker'].unique():
    tick_df = df[df['Ticker']==tick]
    plt.plot(tick_df['Date'], tick_df['Volume'], label=tick)
plt.title('Volume movement for 4 companies', pad=10)
plt.xlabel('Date')
plt.ylabel('Share Units')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

*Apple moves a lot of shares follwoed by Google and Microsoft; Netflix shows less movement.*

## Hypothesis: Apple, Microsoft and Google show similar volume movements

```python
### correlation and p-value check on volume
corr_dict = dict()
tickers = df['Ticker'].unique()
for tick in tickers:
    corr_dict[tick] = df[df['Ticker']==tick]['Volume'].tolist()
corr_df = pd.DataFrame(corr_dict)

print('Ticker1   Ticker2   Correlation   p-value   statistically significant:')
print()
for i in range(len(tickers)-1):
    tick1 = tickers[i]
    for j in range(i+1, len(tickers)):
        tick2 = tickers[j]
        corr, p_v = pearsonr(corr_df[tick1], corr_df[tick2])
        print('{0}   {1}   {2}   {3}   {4}'.format(tick1, tick2, corr, p_v, p_v<0.05))
```

*The result so found is statistically significant.* <br>
**Therefore, Apple, Microsoft and Google show a strong relation in terms of volume of shares movement. Especially Microsoft and Google.**

## Company total market capital movement

```python
### based on close rice * volume
avg_cap_movement = dict()
for tick in df['Ticker'].unique():
    tick_df = df[df['Ticker']==tick]
    plt.plot(tick_df['Date'], tick_df['Close']*tick_df['Volume']/1000000000, label=tick)
    avg_cap_movement[tick] = (tick_df['Close']*tick_df['Volume']/1000000000).mean()
plt.title('Total capital movement for 4 companies', pad=10)
plt.xlabel('Date')
plt.ylabel('Capital (in billions)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
print('Average capital movement by company:')
print()
print('Ticker   Avg capital movement (billions)')
for k, v in avg_cap_movement.items():
    print('{0}   {1}'.format(k, v))
```

*Apple and Microsoft lead in the amount of capital movement in the market.*

## Complete volume movement

```python
df_grouped = df.groupby(df['Date'])['Volume'].sum().reset_index()
df_grouped.columns = ['Date', 'Total Volume']
```

```python
df_grouped['Total Volume'] = df_grouped['Total Volume'] / 1000000
```

```python
plt.plot(df_grouped['Date'], df_grouped['Total Volume'])
plt.title('Total Volume shares of 4 companies', pad=10)
plt.xlabel('Date')
plt.ylabel('Total Shares (count in millions)')
plt.xticks(rotation=90)
plt.show()
```

## Complete capital movement

```python
capital_data = dict()
capital_data['Date'] = df['Date'].tolist()
capital_data['Ticker Total Capital'] = df['Close']*df['Volume']/1000000000
```

```python
df_capital = pd.DataFrame(capital_data)
```

```python
df_grouped = df_capital.groupby(df_capital['Date'])['Ticker Total Capital'].sum().reset_index()
df_grouped.columns = ['Date', 'Total Capital']
```

```python
plt.plot(df_grouped['Date'], df_grouped['Total Capital'])
plt.title('Total Capital movement of 4 companies', pad=10)
plt.xlabel('Date')
plt.ylabel('Total Capital (count in billions)')
plt.xticks(rotation=90)
plt.show()
```

## Volume, Price correlation

```python
for tick in df['Ticker'].unique():
    tick_df = df[df['Ticker']==tick]
    column1 = tick_df['Close']
    column2 = tick_df['Volume']
    dates = tick_df['Date']

    fig, ax1 = plt.subplots()

    ax1.plot(dates, column1, color='blue', label='Close Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Close Price', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    ax2 = ax1.twinx()
    ax2.plot(dates, column2, color='green', label='Volume')
    ax2.set_ylabel('Volume', color='green')
    ax2.tick_params(axis='y', labelcolor='green')

    ax1.tick_params(axis='x', rotation=90)

    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

    plt.title('Close Price and Volume movement for '+tick)
    plt.show()
    print()
    print('Correlation between volume and price for {0}: {1}'.format(tick, tick_df['Volume'].corr(tick_df['Close'])))
```
